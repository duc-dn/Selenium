```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("covid.csv", index_col=0)
data 
```

```python
data.info()
```

```python
plt.figure(figsize=(15, 7))
plt.xticks(rotation ='vertical')
plt.title("Tổng số ca nhiễm covid của các thành phố")
sns.set_style('darkgrid')
sns.barplot(x="Thành phố", y="Tổng số ca", data=data)
```

```python
plt.figure(figsize=(15, 7))
plt.xticks(rotation ='vertical')
sns.set_style('darkgrid')
data["Tỉ lệ tử vong"] = (data["Tử vong"] / data["Tổng số ca"]) * 100
sns.lineplot(x="Thành phố", y="Tỉ lệ tử vong" , data=data)
```
